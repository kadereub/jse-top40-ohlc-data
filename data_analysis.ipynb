```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
current_path = os.getcwd()
data_path = os.path.join(current_path, "_data")
```

```python
data_files = os.listdir(data_path)
string = "6M-ticks-timebars"
```

```python
files = [f for f in data_files if string in f]
```

**Read in Stonks data**

```python
stonks = {}
for f in files:
    # Don't read in STEFI
    if "Index" not in f:
       # Get dict by ticker and assign ohlc data
        stonks[f[0:3]] = pd.read_csv(os.path.join(data_path, f))
```

**Plot Woolworths Close Price**

```python
data = stonks["WHL"].copy()
data.date = pd.to_datetime(data.date)

fig, ax = plt.subplots(nrows=2, sharex=True, figsize=(15,8))

ax[0].plot(data.date, data.close)
ax[1].bar(data.date, data.volume)

# xfmt = mpl.dates.DateFormatter('%H:%M')
# ax[1].xaxis.set_major_locator(mpl.dates.HourLocator(interval=3))
# ax[1].xaxis.set_major_formatter(xfmt)

# ax[1].xaxis.set_minor_locator(mpl.dates.HourLocator(interval=1))
# ax[1].xaxis.set_minor_formatter(xfmt)

ax[1].get_xaxis().set_tick_params(which='major', pad=25)

fig.autofmt_xdate()
plt.show()
```
